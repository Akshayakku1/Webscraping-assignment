# Answer for question 4

```python
page3=requests.get('https://bookpage.com/')
page3
```

```python
page3.content
```

```python
soup3=BeautifulSoup(page3.content)
soup3
```

```python
bookname=soup3.find_all('h4',class_="italic")
```

```python
bookname
```

```python
bookname_name=[]
for l in bookname:
    bookname_name.append(l.text.replace('\n',''))
bookname_name
```

```python
len(bookname_name)
```

```python
authorname=soup3.find_all('p',class_="sans bold")
```

```python
authorname
```

```python
authorname_name=[]
for m in authorname:
    authorname_name.append(m.text.replace('\n',''))
authorname_name
```

```python
len(authorname_name)
```

```python
genre=soup3.find_all('p',class_="genre-links hidden-phone")
genre
```

```python
genre_name=[]
for n in genre:
    genre_name.append(n.text.replace('\n',''))
genre_name
```

```python
len(genre_name)
```

```python
reviews=soup3.find_all('p',class_="excerpt")
reviews
```

```python
reviews_name=[]
for o in reviews:
    reviews_name.append(o.text.replace('\n',''))
reviews_name
```

```python
len(reviews_name)
```

```python
import pandas as pd
books = pd.DataFrame({})
books['bookname']= bookname_name
books['authorname']=authorname_name
books['genre']=genre_name
books['reviews']=reviews_name
books.head(6)
```
